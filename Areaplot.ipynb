```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('mobile_sales.csv')
Sd=df.head(10)
Sd
```

```python
Sd_pivot = Sd.pivot(index='Date', columns='Brand', values='TotalRevenue')
Sd['Date'] = pd.to_datetime(Sd['Date'])
monthly_grouped = Sd.groupby(Sd['Date'].dt.to_period('M')).sum(numeric_only=True)
print(monthly_grouped.head())
# total_revenue_by_date = Sd.groupby('Date')['TotalRevenue'].sum()
```

```python
Sd_pivot.plot(kind='area', alpha=0.5, figsize=(12, 6))
plt.title('Monthly Revenue by Brand')
plt.xlabel('Month')
plt.ylabel('TotalRevenue')
plt.legend(title='Brand')
plt.show()
```

```python
df_pivot = df.pivot(index='Month', columns='Units_sold', values='Total_Revenue')
```

```python
df_pivot.plot(kind='area', alpha=0.5, figsize=(12, 6))
plt.title('Monthly Revenue by Brand')
plt.xlabel('Month')
plt.ylabel('Total_Revenue')
plt.legend(title='Units_sold')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='TotalRevenue', data=Sd, estimator=sum, ci=None)
plt.title('Total Revenue by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Revenue')
plt.show()
```

```python
units_sold = Sd.groupby('Brand')['UnitsSold'].sum()
plt.figure(figsize=(8, 8))
units_sold.plot(kind='pie', autopct='%1.1f%%')
plt.title('Units Sold by Brand')
plt.ylabel('')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
total_revenue_by_date = Sd.groupby('Date')['TotalRevenue'].sum()
plt.figure(figsize=(10, 6))
total_revenue_by_date.plot(kind='line')
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
df.set_index('MobileModel', inplace=True)
```

```python
Sd_model=df.loc[['dinner','quality','direction'],:]
Sd_model
```

```python
df.set_index('MobileModel', inplace=True)
```

```python
total_values = Sd_model['UnitsSold'].sum()
category_proportions = Sd_model['UnitsSold'] / total_values

# print out proportions
pd.DataFrame({"Category Proportion": category_proportions})
```

```python
width = 40 # width of chart
height = 10 # height of chart

total_num_tiles = width * height # total number of tiles

print(f'Total number of tiles is {total_num_tiles}.')
```

```python
tiles_per_category = (category_proportions * total_num_tiles).round().astype(int)

# print out number of tiles per category
pd.DataFrame({"Number of tiles": tiles_per_category})
```

```python
waffle_chart = np.zeros((height, width), dtype = np.uint)

# define indices to loop through waffle chart
category_index = 0
tile_index = 0

# populate the waffle chart
for col in range(width):
    for row in range(height):
        tile_index += 1

        # if the number of tiles populated for the current category is equal to its corresponding allocated tiles...
        if tile_index > sum(tiles_per_category[0:category_index]):
            # ...proceed to the next category
            category_index += 1

        # set the class value to an integer, which increases with class
        waffle_chart[row, col] = category_index

print ('Waffle chart populated!')
```

```python
waffle_chart
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()
plt.show()
```

```python
# instantiate a new figure object
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)

# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])
plt.show()
```

```python
fig = plt.figure()

# use matshow to display the waffle chart
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# get the axis
ax = plt.gca()

# set minor ticks
ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
ax.set_yticks(np.arange(-.5, (height), 1), minor=True)

# add gridlines based on minor ticks
ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

plt.xticks([])
plt.yticks([])

# compute cumulative sum of individual categories to match color schemes between chart and legend
values_cumsum = np.cumsum(Sd_model['UnitsSold'])
total_values = values_cumsum[len(values_cumsum) - 1]

# create legend
legend_handles = []
for i, category in enumerate(Sd_model.index.values):
    label_str = category + ' (' + str(Sd_model['UnitsSold'][i]) + ')'
    color_val = colormap(float(values_cumsum[i])/total_values)
    legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

# add legend to chart
plt.legend(handles=legend_handles,
           loc='lower center',
           ncol=len(Sd_model.index.values),
           bbox_to_anchor=(0., -0.2, 0.95, .1)
          )
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud imported!')
```

```python
import urllib


mobile_sales = urllib.request.urlopen('https://www.kaggle.com/datasets/waqi786/mobile-sales-dataset').read().decode("utf-8")
```

```python
stopwords = set(STOPWORDS)
```

```python
mob_wc = WordCloud()

# generate the word cloud
mob_wc.generate(mobile_sales)
```

```python
plt.imshow(mob_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(14, 18))

# display the cloud
plt.imshow(mob_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
